```python
# missing director values filled with 'Not available'
df['director']=df['director'].fillna('Not Available')

# missing cast values filled with 'Not available'
df['cast']=df['cast'].fillna('Not Available')

# country column filled with most frequent value
frequent_country=df['country'].mode()[0]
df['country']=df['country'].fillna(frequent_country)

# date added filled with most frequent value
frequent_date=df['date_added'].mode()[0]
df['date_added']=df['date_added'].fillna(frequent_date)

# rating column filled with most frequent value
frequent_rating=df['rating'].mode()[0]
df['rating']=df['rating'].fillna(frequent_rating)

# duration filled with most frequent value
frequent_duration=df['duration'].mode()[0]
df['duration']=df['duration'].fillna(frequent_duration)
```

```python
## Correlation between Genre and Country
```

```python
country_df=df.copy()
country_df['listed_in'] = country_df['listed_in'].astype('str').apply(
    lambda x: x.split(', '))

# Explode the rows with multiple categories to different rows
country_genre_df = country_df.explode('listed_in').drop_duplicates()
```

```python
#Top 20 countries by count of titles
top_20 = country_df.country.value_counts().reset_index()
top_20 = top_20.loc[:20,'index'].to_list()
#top_20.remove('nan')
top_20_df = country_genre_df.loc[country_genre_df['country'].isin(top_20)]
```

```python
def custom_agg(x):
    return x.nunique()

data = pd.pivot_table(top_20_df, index = top_20_df['listed_in'],
                      columns = top_20_df['country'],
                      values = 'show_id', aggfunc=custom_agg)
```

```python
plt.figure(figsize=(12,8))
#ax=sns.heatmap(data,annot=True,fmt=".0f")
ax = sns.heatmap(data, annot=True, fmt=".0f",
                  norm=matplotlib.colors.LogNorm(vmin=data.min().min(),
                                                 vmax=data.max().max()))
ax.set(xlabel="Countries", ylabel="Categories", title='Countries v/s Genres');
```

```python
Insight: Action & Adventure, Children and Family movies,Comedies are popular genre
    
```
